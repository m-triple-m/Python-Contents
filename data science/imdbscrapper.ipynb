```python
from bs4 import BeautifulSoup
import requests
```

```python
data = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
```

```python
if data.status_code == 200:
    print('success')
elif data.status_code == 404:
    print('page not found')
elif data.status_code == 500:
    print('server error')
```

```python
soup = BeautifulSoup(data.text)
```

```python
rows = soup.find('tbody').findAll('tr')
```

```python
len(rows)
```

```python
for movie in rows:
    print(movie.find('td', {'class' : 'titleColumn'}).find('a').text)
    print(movie.find('td', {'class' : 'titleColumn'}).find('span').text.strip('()'))
    print(movie.find('td', {'class' : 'imdbRating'}).find('strong').text)
```

```python
movie_details = []

for movie in rows:
    detail = {}
    detail['name'] = movie.find('td', {'class' : 'titleColumn'}).find('a').text
    detail['year'] = movie.find('td', {'class' : 'titleColumn'}).find('span').text.strip('()')
    detail['rating'] = movie.find('td', {'class' : 'imdbRating'}).find('strong').text

    movie_details.append(detail)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(movie_details)
```
