```python
import requests
from bs4 import BeautifulSoup
```

```python
data = requests.get("https://www.flipkart.com/search?q=backpack")
```

```python
soup = BeautifulSoup(data.text)
```

```python
row1 = soup.findAll('div', {'class' : '_1xHGtK _373qXS'})
```

```python
len(row1)
```

```python
for bagdetail in row1:
    print('www.flipkart.com'+bagdetail.find('a').attrs.get('href'))
    print(bagdetail.find('div', {'class' : '_2WkVRV'}).text)
    print(bagdetail.find('div', {'class' : '_3eWWd-'}).text)
    print(bagdetail.find('div', {'class' : '_3Ay6Sb'}).find('span').text)
    print('')
```
